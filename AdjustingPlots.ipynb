```
import altair as alt
import numpy as np
import pandas as pd
```

```
from vega_datasets import data
df = data.countries()
df.head()
df2000 = df[df.year==2000]
```

```
df.tail()
```

```
df2000 = df[df.year==2000]
df2000.head()
```

```
alt.Chart(df2000).mark_point().encode(
    x='country:N',
    y='life_expect:Q'
)
```

```
alt.Chart.configure_point_
```

```
alt.Chart(df2000).mark_point().encode(
    x='country:N',
    y='life_expect:Q'
).configure_point(
    size=100,
    color='red',
    filled=True
)
```

```
alt.Chart(df2000).mark_point(color = 'blue', filled = False).encode(
    x='country:N',
    y='life_expect:Q'
).configure_point(
    size=100,
    color='red',
    filled=True
)
```

```
alt.Chart(df2000).mark_point(color = 'blue', filled = False).encode(
    x='country:N',
    y='life_expect:Q',
    color = alt.value('green')
).configure_point(
    size=100,
    color='red',
    filled=True
)
```

```
alt.Chart(df2000).mark_point().encode(
  #  x='country:N',
  #  y='life_expect:Q',
    x = alt.X('country:N', title = 'Country'),
    y = alt.Y('life_expect', title = 'Life expectancy in year 2000')
).configure_point(
    size=100,
    color='red',
    filled=True
)
```

```
alt.Chart(df2000).mark_point().encode(
  #  x='country:N',
  #  y='life_expect:Q',
    x = alt.X('country:N', title = 'Country'),
    y = alt.Y('life_expect', title = 'Life expectancy in year 2000')
).configure_point(
    size=100,
    color='red',
    filled=True
).properties(
title = ' Country vs Life Expectancy of People Living there in 2000')
```

```
alt.Chart(df2000).mark_bar().encode(
    x=alt.X('country', bin=False, axis=alt.Axis(labelAngle=90)),
    y=alt.Y('life_expect', axis=alt.Axis(labels=True, ticks=False, title='Life Expectancy in 2000'))
)
```

```
alt.Chart(df2000).mark_bar().encode(
    x=alt.X('country', bin=False, axis=alt.Axis(labelAngle=90)),
    y=alt.Y('life_expect', axis=alt.Axis(labels=True, ticks=False, title='Life Expectancy in 2000'))
).configure_axisY(
labelAngle = 45)
```

```
alt.Chart(df2000).mark_bar().encode(
    x=alt.X('country', scale = alt.Scale(domain=['India','Pakistan','United States','Ireland'])),
    y=alt.Y('life_expect', scale=alt.Scale(domain=[20, 90])),
)
```
